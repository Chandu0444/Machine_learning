#LDA

```python
import pandas as pd
df = pd.DataFrame({"Gender":["Male","Male","Male","Male","Male","Male","Female","Female","Female","Female","Female","Female"],
                   "Shoe_Size":[9,8,8.5,10,7,9.5,7,6,6.5,5,8,7.5],
                   "Height":[6.1,5.7,5.4,5.8,6.2,5.6,5.5,5.1,5.2,5,4.11,5.9]})
df
```

```python
df_male=df[df['Gender']=='Male']
df_female=df[df['Gender']=='Female']
```

```python
m_array = df_male[["Shoe_Size","Height"]].to_numpy().transpose()
f_array = df_female[["Shoe_Size","Height"]].to_numpy().transpose()
```

```python
import numpy as np
Cov_male = np.cov(m_array)
Cov_male
```

```python
Cov_female = np.cov(f_array)
Cov_female
```

```python
w= (Cov_male + Cov_female)/2
w
```

Find inverse for W

```python
w_inv=np.linalg.inv(w)
w_inv
```

```python
total_array=df[["Shoe_Size","Height"]].to_numpy().transpose()
```

```python
cov_t=np.cov(total_array)
cov_t
```

```python
B = cov_t-w
B
```

```python
# step1: find the seperation matrix

S=np.dot(w_inv,B)
S
```

```python
# Step-2
# Find the eigen Value and Eigen Vector for S matrix

eig_vals =np.linalg.eig(S)
eig_vals
```

```python
# Step-3
# Sort the eigen value from higher to lower, find the eigen vector corresponding to highest eigen value
e = np.array([0.4690,0.8831]) #--> eigen vector corresponding to eigen value 1.711
```

```python
# step4: finding linear discriminent
total_array
```

```python
discrimnent=np.dot(e,total_array)
```

```python
df['descrimenent']=np.dot(e,total_array)
df
```

#PCA

```python
from sklearn.datasets import load_wine
from sklearn.preprocessing import StandardScaler
```

```python
data=load_wine()
```

```python
data
```

```python
x=data['data']
scalar=StandardScaler()
x_standrdzed=scalar.fit_transform(x)
```

```python
y=data['target']
```

```python
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
```

```python
pca= PCA(n_components=2)
lda=LinearDiscriminantAnalysis(n_components=2)
```

```python
x_reduced_byPCA=pca.fit_transform(x_standrdzed)
x_reduced_byLDA=lda.fit_transform(x_standrdzed,y) # for lda we should give both x and y
```

#Classifier using PCA reduced dim

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.linear_model import LogisticRegression
```

```python
x_train, x_test, y_train, y_test= train_test_split(x_reduced_byPCA,y,test_size=0.3)
log1=LogisticRegression()
log_model1=log1.fit(x_train,y_train)
y_test_pred=log_model1.predict(x_test)
print(confusion_matrix(y_test_pred,y_test))
```

#classifier using LDA reduced dim

```python
x_train_lda, x_test_lda, y_train_lda , y_test_lda =  train_test_split(x_reduced_byLDA,y, test_size=0.3)
log2=LogisticRegression()
log_model2=log2.fit(x_train,y_train)
y_test_pred1=log_model2.predict(x_test)
print(confusion_matrix(y_test_pred1,y_test))
```
