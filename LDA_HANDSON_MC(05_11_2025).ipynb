LDA  handson with digit data set

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn import datasets
```

```python
data=datasets.load_digits()
```

```python
data
```

```python
x=data['data']
y=data['target']
```

Another method for standardization, this method saves the time

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train, y_test= train_test_split(x,y,test_size=0.30)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
```

```python
x_train_sc=sc.fit_transform(x_train)
x_test_sc=sc.transform(x_test)
```

```python
x_train_sc.shape
```

```python
x_test_sc.shape
```

#Implementing the LDA, for dimensionality reducttion.

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
```

lda=LinearDiscriminantAnalysis(n_components=?)

how to find "n_components"

no.of classes:- {0,1,2,3,...9} = 10

no.of features:- 64

min(no.of classes - 1 , no.of features)

min(64,10-1) = min(64,9)

finally n_component is :- 9


```python
lda=LinearDiscriminantAnalysis(n_components=9)
x_train_lda=lda.fit_transform(x_train_sc,y_train)
x_train_lda.shape
```

64 columns reduced to 9 columns

```python
x_test_lda=lda.transform(x_test_sc) # here we dont need specify the fit cause already dataset was fit and tranformed above^
x_test_lda.shape
```

```python
# models: log reg, d tree, random forest.
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import *
from sklearn.model_selection import train_test_split
```

```python
#Logistic regression
lg=LogisticRegression()
model_lg=lg.fit(x_train_lda,y_train)
y_pred1= model_lg.predict(x_test_lda)
accuracy_score(y_test,y_pred1)*100
```

```python
#D tree
dt=DecisionTreeClassifier()
model_dt=dt.fit(x_train_lda,y_train)
y_pred2= model_dt.predict(x_test_lda)
accuracy_score(y_test,y_pred2)*100
```

```python
#RandomForestClassifier
rd=RandomForestClassifier(n_estimators=100)
model_rd=rd.fit(x_train_lda,y_train)
y_pred3= model_rd.predict(x_test_lda)
accuracy_score(y_test,y_pred3)*100
```
