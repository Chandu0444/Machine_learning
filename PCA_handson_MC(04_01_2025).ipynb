PCA handson with digits dataset.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn import datasets
```

```python
data=datasets.load_digits()
```

```python
data
```

#splitting the dataset as x,y variables

```python
x=data['data']
y=data['target']
```

```python
x
```

```python
y
```

Standardise the data.(all values get closer the mean)

```python
from sklearn.preprocessing import StandardScaler
```

```python
se=StandardScaler()
x_std=se.fit_transform(x)
x_std
```

```python
x_std.shape     #64 feature
```

Implementing the pca for dimensionality(features) reduction.

we have 64 columns

```python
from sklearn.decomposition import PCA
```

```python
pca=PCA(n_components=0.80)
x_pca=pca.fit_transform(x_std)
print(f"No.of rows and cols after implementing the the pca: {x_pca.shape}")
x_pca
```

Using pca we reduced 64 columns **to** 21 columns

Models that are suitable for classification - log reg, decion tree, random forest.


```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
#log reg
le=LogisticRegression()
x_train_pca, x_test_pca, y_train, y_test=train_test_split(x_pca,y,test_size=0.2)
model_fit1=le.fit(x_pca,y)
y_test_pred1=model_fit1.predict(x_test_pca)
accuracy_score(y_test,y_test_pred1)*100
```

```python
#decision tree
from sklearn.tree import DecisionTreeClassifier
de=DecisionTreeClassifier()
model_fit2=de.fit(x_train_pca,y_train)
y_test_pred2=model_fit2.predict(x_test_pca)
accuracy_score(y_test,y_test_pred2)*100
```

```python
#random forest
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier(n_estimators=100)
model_fit3=rf.fit(x_train_pca,y_train)
y_test_pred3=model_fit3.predict(x_test_pca)
accuracy_score(y_test_pred3,y_test)*100
```
