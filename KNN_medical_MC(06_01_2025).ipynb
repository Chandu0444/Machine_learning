KNN (K-Nearest Neighbours)
it predicts a label for a new data point by looking at its neighbors.
STEPS

1. Pick a K value (K=number of neighbors)

2. For a new data point(called as query point), calculate its distance(Euclidean distance) from all other points in the dataset.

3. Find the K closest points(neighbors)

4. Classify on the basis of majority

```python
# removing columns which has mor than 50% of the null values
data_new = data[['Disease', 'Symptom_1','Symptom_2',	'Symptom_3', 'Symptom_4', 'Symptom_5', 'Symptom_6']]
```

```python
data_new
```

```python
for col in data_new.columns:
  if data_new[col].dtype=='object':
    data_new[col] = data_new[col].fillna(data_new[col].mode()[0])
  else:
    data_new[col] = data_new[col].fillna(data_new[col].mean())
```

```python
data_new['Disease'].mode()[1]
```

```python
data_new.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
for col in data_new.columns:
    data_new[col] = le.fit_transform(data_new[col])

    for idx, cls in enumerate(le.classes_):
      print(f"{cls}->{idx}")
```

# In KNN we do not remove duplicates, because the duplicates strengthens the classification

```python
data_new['Disease'].values
```

#Clustering Algo

```python
x=data_new.drop('Disease',axis=1)
y=data_new['Disease']
```

```python
y.unique()
```

```python
y.nunique()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.20, random_state=52)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
knn.fit(x_train,y_train)
```

```python
y_pred = knn.predict(x_test)
```

```python
from sklearn.metrics import *
acc=accuracy_score(y_test,y_pred)*100
acc
```

 What is .score?

 In machine learning, .score is a method in libraries like scikit-learn that returns a quantitative metric evaluating how well a trained model fits the data.

```python
acc2 = knn.score(x_train,y_train)
acc2
```
