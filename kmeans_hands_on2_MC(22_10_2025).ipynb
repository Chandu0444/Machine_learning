```python
data_copy=data.copy()
```

```python
data_copy.drop('Player',axis=1,inplace=True)
```

```python
data_copy.info()
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
```

```python
data_scaled=sc.fit_transform(data_copy)
data_scaled
```

```python
data_scaled=pd.DataFrame(data_scaled,columns=data_copy.columns)
data_scaled
```

```python
from sklearn.cluster import KMeans
```

```python
k_value=[2,3,4,5,6,7,8,9,10]
```

```python
wcss=[]
for k in k_value:
  model=KMeans(n_clusters=k,max_iter=100,random_state=25)
  model.fit(data_scaled)
  wcss.append(model.inertia_)
```

```python
plt.plot(k_value,wcss)
```

```python
final_model=KMeans(n_clusters=3,max_iter=100,random_state=25)
final_model.fit(data_scaled)
```

```python
final_model.labels_
```

```python
data['ClusterID']=final_model.labels_
data
```

```python
group1=data[data['ClusterID']==0]
group1
```

```python
group2=data[data['ClusterID']==1]
group2
```
